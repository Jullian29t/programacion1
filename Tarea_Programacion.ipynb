**LISTA**

```python
lista = ['J','U','U','U','L','I','A','N']
print(lista)
tupla = tuple(lista)
print(tupla)
diccionario = dict(enumerate(lista))
print(diccionario)
conjunto = set(lista)
print(conjunto)
```

**TUPLA**

```python
tupla2 = ('J','U','U','U','L','I','A','N')
print(tupla2)
lista2 = list(tupla2)
print(lista2)
diccionario2 = dict(enumerate(tupla2))
print(diccionario2)
conjunto2 = set(tupla2)
print(conjunto2)
```

**DICCIONARIO**

```python
diccionario3 = {'nombre':'Julian','apellido':'usma'}
print(diccionario3)
lista3 = list(diccionario3.values())+list(diccionario3)
print(lista3)
tupla3 = tuple(diccionario3.values())+tuple(diccionario3)
print(tupla3)
conjunto3 = set(list(diccionario3.values())+list(diccionario3))
print(conjunto3)
```

**CONJUNTO**

```python
conjunto4 = {'J','U','U','U','L','I','A','N'}
print(conjunto4)
tupla4 = tuple(conjunto4)
print(tupla4)
diccionario4 = dict(enumerate(conjunto4))
print(diccionario4)
lista4 = list(conjunto4)
print(lista4)
```
